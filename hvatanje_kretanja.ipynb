#### Ovde treba tekst za kod dole, odnosno kako i sta se desava dole, objasnjenje itd, nek bude isto formatiran tekst da ne izgleda bezveze, nisam siguran da li treba da se ispisu one silne jednacine

```python
import numpy as np
import math 
from pylab import *
g = 9.81
print('Unesite nagib brda u stepenima:') 
nagib = float(input())
t1 = 0
v1 = 0
r1 = 0
print('Unesite vreme koje je proslo u sekundama:') 
t2 = float(input())


a = g * math.sin(math.radians(nagib))
dt = t2 - t1
v2 = a * dt
r2 = v2 * dt

x=math.cos(math.radians(nagib))*r2
y=-math.sin(math.radians(nagib))*r2
print('Brzina: ', v2, 'Predjeni put: ', r2)

plot(0, 0, 'o')
axis('equal')
xlabel('x [m]')
ylabel('y [m]')
title('Polozaj automobila na brdu nakon %.2f sekundi' % t2)
plot(x, y, 'o')
quiver(x, y, angles = 'xy', scale_units = 'xy', scale = 1)
plot([], [], ' ', label="Brzina: %f$m/s^2$" % v2)
plot([], [], ' ', label="Predjeni put: %fm" % r2)
legend()
show()
```

```python
# Motion Capture (180)
# Motion 1
from pylab import *
# g = 9.81;
t,ax,ay = loadtxt('motion1.d', usecols=[0,1,2], unpack = True);
n = len(t);
a = zeros((n, 2), float);
a[:, 0] = ax;
a[:, 1] = ay;
r = zeros((n, 2), float);
v = zeros((n, 2), float);
n_v = zeros((n, 1), float);
r[0,:] = array([0.0, 0.0]);
v[1,:] = array([0.0, 0.0]);
n_v[0] = 0;
dt = t[1] - t[0];
for i in range(n-1):
    v[i+1, :] = v[i,:] + a[i,:]*dt;
    r[i+1, :] = r[i,:] + v[i,:]*dt;
    n_v[i+1] = norm(v[i+1,:]);
xlabel('x [m]'); ylabel('y [m]'); title("Polozaji automobila (Motion 1)"); axis('equal');
plot(r[:, 0], r[:, 1], '-k');
show();
xlabel('t [s]'); ylabel('v [m/s]'); title("Promena brzine u odnosu na vreme kretanja automobila");
plot(t, n_v, '-k');
show();
```
