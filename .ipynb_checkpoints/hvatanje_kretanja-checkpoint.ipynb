## Hvatanje kretanja

U ovom projektu bavićemo se kretanjem tela i predstavljanjem tog kretanja kroz određene fizičke veličine. Za početak, posmatraćemo automobil koji se nalazi na brdu, pri čemu ćemo zanemariti masu samog automobila.
Zamislimo da se automobil kreće niz beskonačno brdo. Znamo koliki je nagib brda i koliko se dugo automobil kreće. Potrebno je odrediti brzinu i pređeni put, odnosno položaj automobila na brdu nakon predviđenog vremena. Rešenje problema predstavili smo u kodu ispod.
Korisnik unosi nagib brda i vreme kretanja. Pošto je telo u početnom položaju u stanju mirovanja, početna brzina je 0m/s. Ubrzanje računamo po formuli a=g\*sin(nagib). Brzinu v2 računamo na osnovu formule vektora prosečnog ubrzanja a=(v(t+dt)-v(t))/dt, pri čemu je v(t) u kodu označeno kao v1=0, što je početna brzina, a v(t+dt) kao v2, dok dt predstavlja razliku željenog vremena t2 i početnog vremena t1=0, te dolazimo do formule v2=a\*dt. Isti princip je primenjen i u traženju pređenog puta r2 po formuli vektora prosečne brzine.
Kada smo odredili tražene fizičke veličine, predstavićemo sliku u koordinatnom sistemu. Određujemo x i y koordinate na osnovu trigonometrijskog trougla, gde je pređeni put hipotenuza, x koordinata nalegla, a y naspramna kateta.

```python
import numpy as np
import math 
from pylab import *
g = 9.81
print('Unesite nagib brda u stepenima:') 
nagib = float(input())
t1 = 0
v1 = 0
r1 = 0
print('Unesite vreme koje je prošlo u sekundama:') 
t2 = float(input())

a = g * math.sin(math.radians(nagib))
dt = t2 - t1
v2 = a * dt
r2 = v2 * dt

x=math.cos(math.radians(nagib))*r2
y=-math.sin(math.radians(nagib))*r2
print('Brzina: {}m/s'.format(v2), 'Pređeni put: {}m'.format(r2))

plot(0, 0, 'o')
axis('equal')
xlabel('x [m]')
ylabel('y [m]')
title('Položaj automobila na brdu nakon %.2f sekundi' % t2)
plot(x, y, 'o')
quiver(x, y, angles = 'xy', scale_units = 'xy', scale = 1)
plot([], [], ' ', label='Brzina: %fm/s' % v2)
plot([], [], ' ', label='Pređeni put: %fm' % r2)
legend()
show()
```

Naredni problem se ne razlikuje naročito od prethodnog, te ni sam kod. Naime, sada znamo i visinu na kojoj se automobil nalazi. Ona će predstavljati vrh brda, polaznu tačku sa koje automobil kreće. Jedina razlika je u tome što će kod predstavljanja slike u koordinatnom sistemu, polazna tačka biti na koordinatama x=0 i y=h, umesto u koordinatnom početku kao na prethodnoj slici.

```python
import numpy as np
import math 
from pylab import *
g = 9.81
print('Unesite nagib brda u stepenima:') 
nagib = float(input())
t1 = 0
v1 = 0
r1 = 0
print('Unesite vreme koje je prošlo u sekundama:') 
t2 = float(input())
print('Unesite visinu u metrima na kojoj se automobil nalazi:')
h = float(input())

a = g * math.sin(math.radians(nagib))
dt = t2 - t1
v2 = a * dt
r2 = v2 * dt

x=math.cos(math.radians(nagib))*r2
y=-math.sin(math.radians(nagib))*r2
print('Brzina: {}m/s'.format(v2), 'Pređeni put: {}m'.format(r2))

plot(0, h, 'o')
axis('equal')
xlabel('x [m]')
ylabel('y [m]')
title('Položaj automobila na brdu nakon %.2f sekundi' % t2)
plot(x, y+h, 'o')
quiver(0, h, x, y+h-h, angles = 'xy', scale_units = 'xy', scale = 1)
plot([], [], ' ', label="Brzina: %fm/s" % v2)
plot([], [], ' ', label="Pređeni put: %fm" % r2)
legend()
show()
```

U narednom kodu princip problema je isti, ali je razlika u tome što sada ne posmatramo "beskonačno" brdo, već za podnožje brda uzimamo x osu koordinatnog sistema. Problem delimo na dva scenarija; kada se automobil nakon unetog vremena kretanja i dalje nalazi na brdu ili tačno u njegovom podnožju i kada nastavlja da se kreće pravolinijski po ravnoj površini.
U prvom slučaju, kod i problem su isti kao i prethodni.
U drugom slučaju, potrebne su nam nove vrednosti fizičkih veličina. Prvo određujemo pređeni put na brdu, odnosno dužinu puta od vrha do podnožja brda, što računamo na osnovu trigonometrijskog trougla kako znamo visinu brda h, što će predstavljati naspramnu katetu, kao i nagib brda. Zatim određujemo x koordinatu podnožja, pri čemu će y koordinata biti 0. Nakon toga određujemo proteklo vreme potrebno automobilu da se spusti niz brdo na osnovu formula vektora prosečnog ubrzanja i brzine. Zatim, na osnovu ubrzanja i do sada proteklog vremena određujemo brzinu automobila u podnožju koju smo označili sa v2novo. Pošto automobil više nije na brdu, na njega ne utiče ubrzanje a, pa će brzina tokom preostalog pravolinijskog puta biti jednaka onoj koja je bila u podnožju. Ostatak pravolinijskog puta računamo kao proizvod brzine v2novo i preostalog vremena kretanja (razlika unetog vremena i vremena kretanja potrošenog na brdu).

```python
import numpy as np
import math 
from pylab import *
g = 9.81
print('Unesite nagib brda u stepenima:') 
nagib = float(input())
t1 = 0
v1 = 0
r1 = 0
print('Unesite vreme koje je prošlo u sekundama:') 
t2 = float(input())
print('Unesite visinu brda u metrima:')
h = float(input())

a = g * math.sin(math.radians(nagib))
dt = t2 - t1
v2 = a * dt
r2 = v2 * dt

x=math.cos(math.radians(nagib))*r2
y=-math.sin(math.radians(nagib))*r2

plot(0, h, 'o')
axis('equal')
xlabel('x [m]')
ylabel('y [m]')
title('Položaj automobila na brdu nakon %.2f sekundi' % t2)
if (y+h) >= 0:
    plot(x, y+h, 'o')
    quiver(0, h, x, y+h-h, angles = 'xy', scale_units = 'xy', scale = 1)
    print('Brzina: {}m/s'.format(v2), 'Pređeni put: {}m'.format(r2))   
    plot([], [], ' ', label="Brzina: %fm/s" % v2)
    plot([], [], ' ', label="Pređeni put: %fm" % r2)
    legend()
    show()
else:
    hipotenuza=h/math.sin(math.radians(nagib))
    xpodnozje=math.cos(math.radians(nagib))*hipotenuza
    dt1=sqrt(hipotenuza/a)
    dtostatak=t2-dt1
    v2novo=a*dt1
    plot(xpodnozje, 0, 'o', label="Brzina: %fm/s" % v2novo)
    quiver(0, h, xpodnozje, 0-h, angles = 'xy', scale_units = 'xy', scale = 1)
    plot([], [], ' ', label="Pređeni put brda: %fm" % hipotenuza)
    plot([], [], ' ', label="Vreme kretanja nizbrdo: %fs" % dt1)
    #sada se kreće pravolinijski i nema ubrzanja u odnosu na brzinu u podnožju, odnosno brzina v2novo ostaje ista tokom celog puta
    r2novo=v2novo*dtostatak
    plot(xpodnozje+r2novo, 0, 'o', label="Brzina: %fm/s" % v2novo)
    quiver(xpodnozje, 0, xpodnozje+r2novo-xpodnozje, 0, angles = 'xy', scale_units = 'xy', scale = 1)

    print('Brzina: {}m/s'.format(v2novo), 'Pređeni put: {}m'.format(hipotenuza + r2novo))    
    plot([], [], ' ', label="Pređeni pravolinijski put: %fm" % r2novo)
    plot([], [], ' ', label="Vreme kretanja pravolinijski: %fs" % dtostatak)
    plot([], [], ' ', label="Ukupan pređeni put: %fm" % (hipotenuza + r2novo))
    legend()
    show()
```

```python
# Motion Capture (180)
# Motion 1
from pylab import *
# g = 9.81;
t,ax,ay = loadtxt('motion1.d', usecols=[0,1,2], unpack = True);
n = len(t);
a = zeros((n, 2), float);
a[:, 0] = ax;
a[:, 1] = ay;
r = zeros((n, 2), float);
v = zeros((n, 2), float);
n_v = zeros((n, 1), float);
r[0,:] = array([0.0, 0.0]);
v[1,:] = array([0.0, 0.0]);
n_v[0] = 0;
dt = t[1] - t[0];
for i in range(n-1):
    v[i+1, :] = v[i,:] + a[i,:]*dt;
    r[i+1, :] = r[i,:] + v[i,:]*dt;
    n_v[i+1] = norm(v[i+1,:]);
xlabel('x [m]'); ylabel('y [m]'); title("Polozaji automobila (Motion 1)"); axis('equal');
plot(r[:, 0], r[:, 1], '-k');
show();
xlabel('t [s]'); ylabel('v [m/s]'); title("Promena brzine u odnosu na vreme kretanja automobila");
plot(t, n_v, '-k');
show();
```
