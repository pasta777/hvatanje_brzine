## Motion Capture

In this project, we will deal with the movement of the body and the representation of that movement through certain physical quantities. To begin with, we will consider a car on a hill, neglecting the mass of the car itself.
Let's imagine a car moving down an infinite hill. We know the steepness of the hill and the duration of time for which the car has been moving. The speed and distance traveled needs to be determined, i.e. the position of the car on the hill after the estimated time. We have presented the solution to the problem in the code below.

The user enters the slope of the hill and the travel time. Since the body is at rest in the initial position, the initial velocity is 0 m/s. Acceleration is calculated using the formula *a = g / sin(slope)*. Velocity *v2* is calculated based on acceleration vector formula *a = (v(t + dt) - v(t)) / dt*, where v(t) is marked in the code as *v1 = 0*, which is the initial velocity, and *v(t + dt)* as *v2*, where *dt* represents the difference between the time *t2* and initial time *t1 = 0*, thus we construct the formula *v2 = a \* dt*. The same principle is applied in finding the traveled distance *r2* using the average speed vector formula.

Once we have determined the required physical quantities, we will present the image in the coordinate system. We determine the x and y coordinates on the basis of a trigonometric triangle, where the distance covered is the hypotenuse.

```python
import numpy as np
import math 
from pylab import *
g = 9.81
print('Enter the slope of the hill in degrees:') 
slope = float(input())
t1 = 0
v1 = 0
r1 = 0
print('Enter the elapsed time in seconds:') 
t2 = float(input())

a = g * math.sin(math.radians(slope))
dt = t2 - t1
v2 = a * dt
r2 = v2 * dt

x=math.cos(math.radians(slope))*r2
y=-math.sin(math.radians(slope))*r2
print('Velocity: {}m/s'.format(v2), 'Distance traveled: {}m'.format(r2))

plot(0, 0, 'o')
axis('equal')
xlabel('x [m]')
ylabel('y [m]')
title('The position of the car on the hill after %.2f seconds' % t2)
plot(x, y, 'o')
quiver(x, y, angles = 'xy', scale_units = 'xy', scale = 1)
plot([], [], ' ', label='Velocity: %fm/s' % v2)
plot([], [], ' ', label='Distance traveled: %fm' % r2)
legend()
show()
```

The next problem is not particularly different from the previous one, and neither is the code itself. Namely, now we know the height at which the car is located. The height represents the top of the hill, the starting point from which the car starts. The only difference is that when presenting the image in the coordinate system, the starting point will be at the coordinates *x = 0* and *y = h*, instead of at the coordinate origin as in the previous image.

```python
import numpy as np
import math 
from pylab import *
g = 9.81
print('Enter the slope of the hill in degrees:') 
slope = float(input())
t1 = 0
v1 = 0
r1 = 0
print('Enter the elapsed time in seconds:') 
t2 = float(input())
print('Enter the height in meters where the car is located:')
h = float(input())

a = g * math.sin(math.radians(slope))
dt = t2 - t1
v2 = a * dt
r2 = v2 * dt

x=math.cos(math.radians(slope))*r2
y=-math.sin(math.radians(slope))*r2
print('Velocity: {}m/s'.format(v2), 'Distance travelled: {}m'.format(r2))

plot(0, h, 'o')
axis('equal')
xlabel('x [m]')
ylabel('y [m]')
title('The position of the car on the hill after %.2f seconds' % t2)
plot(x, y+h, 'o')
quiver(0, h, x, y+h-h, angles = 'xy', scale_units = 'xy', scale = 1)
plot([], [], ' ', label="Velocity: %fm/s" % v2)
plot([], [], ' ', label="Distance traveled: %fm" % r2)
legend()
show()
```

In the following code, the principle of the problem is the same, but the difference is that now we are not observing an "infinite" hill, but we are taking the x axis of the coordinate system as the foot of the hill. We divide the problem into two scenarios; when, after the entered travel time, the car is still on the hill or exactly at its foot and continues to move in a straight line on a flat surface.

In the first case, the code and the problem are the same as the previous one.

In the second case, we need new values of physical quantities. First, we determine the distance traveled on the hill, that is, the length of the road from the top to the foot of the hill, which we calculate based on the trigonometric triangle, as we know the height of the hill h, which will represent the opposite leg, as well as the slope of the hill. Next, we determine the x coordinate of the foot, where the y coordinate will be 0. After that, we determine the elapsed time required for the car to go down the hill based on the average acceleration and velocity vector formulas. Then, based on the acceleration and the time elapsed so far, we determine the velocity of the car at the base, which we marked with *v2_new*. Since the car is no longer on the hill, it is not affected by the acceleration a, so the speed for the rest of the straight-line journey will be the same as it was at the bottom. The rest of the straight path is calculated as the product of the *v2_new* velocity and the remaining movement time (the difference between the entered time and the movement time spent on the hill).

```python
import numpy as np
import math 
from pylab import *
g = 9.81
print('Enter the slope of the hill in degrees:') 
slope = float(input())
t1 = 0
v1 = 0
r1 = 0
print('Enter the elapsed time in seconds:') 
t2 = float(input())
print('Enter the height of the hill in meters:')
h = float(input())

a = g * math.sin(math.radians(slope))
dt = t2 - t1
v2 = a * dt
r2 = v2 * dt

x = math.cos(math.radians(slope))*r2
y = -math.sin(math.radians(slope))*r2

plot(0, h, 'o')
axis('equal')
xlabel('x [m]')
ylabel('y [m]')
title('The position of the car after %.2f seconds' % t2)
if (y+h) >= 0:
    plot(x, y+h, 'o')
    quiver(0, h, x, y+h-h, angles = 'xy', scale_units = 'xy', scale = 1)
    print('Velocity: {}m/s'.format(v2), 'Distance traveled: {}m'.format(r2))   
    plot([], [], ' ', label="Velocity: %fm/s" % v2)
    plot([], [], ' ', label="Distance traveled: %fm" % r2)
    legend()
    show()
else:
    hypotenuse = h / math.sin(math.radians(slope))
    x_foot = math.cos(math.radians(slope)) * hypotenuse
    dt1 = sqrt(hypotenuse/a)
    dt_remainder = t2 - dt1
    v2_new = a * dt1
    plot(x_foot, 0, 'o', label="Velocity: %fm/s" % v2_new)
    quiver(0, h, x_foot, 0-h, angles = 'xy', scale_units = 'xy', scale = 1)
    plot([], [], ' ', label="Distance traveled on hill: %fm" % hypotenuse)
    plot([], [], ' ', label="Time traveled downhill: %fs" % dt1)
    # Now it moves in a straight line and there is no acceleration relative to the speed at the foot,
    # i.e. the speed of the v2_new remains the same throughout the journey
    r2_new = v2_new * dt_remainder
    plot(x_foot + r2_new, 0, 'o', label="Velocity: %fm/s" % v2_new)
    quiver(x_foot, 0, x_foot + r2_new - x_foot, 0, angles = 'xy', scale_units = 'xy', scale = 1)

    print('Velocity: {}m/s'.format(v2_new), 'Distance traveled: {}m'.format(hypotenuse + r2_new))    
    plot([], [], ' ', label="Distance traveled in a straight line: %fm" % r2_new)
    plot([], [], ' ', label="Time traveled in a straight line:%fs" % dt_remainder)
    plot([], [], ' ', label="Total distance traveled: %fm" % (hypotenuse + r2_new))
    legend()
    show()
```

According to the graphic of the position of the car, we can see the path, but not the number of passes through the "cycle". However, through the graph of the change of velocity in relation to the time of movement, we can conclude that the car went through the cycle exactly once, because between two linear intervals we can notice an interval that resembles a parabola that indicates a journey through the cycle. It turns out that the acceleration is the highest from the top of the cycle to the extreme left point of the cycle because only the gravitational force acts on the body in that time interval. Also this can also be determined on the graph of the change of velocity in relation to the time of car's movement, where the derivative of the speed by time is the highest around eight second.

```python
# Motion Capture (180)
# Motion 1
from pylab import *
# g = 9.81;
t, ax, ay = loadtxt('motion1.d', usecols = [0,1,2], unpack = True);
n = len(t);
a = zeros((n, 2), float);
a[:, 0] = ax;
a[:, 1] = ay;
r = zeros((n, 2), float);
v = zeros((n, 2), float);
n_v = zeros((n, 1), float);
r[0, :] = array([0.0, 0.0]);
v[1, :] = array([0.0, 0.0]);
n_v[0] = 0;
dt = t[1] - t[0];
for i in range(n - 1):
    v[i + 1, :] = v[i, :] + a[i, :] * dt;
    r[i + 1, :] = r[i ,:] + v[i, :] * dt;
    n_v[i + 1] = norm(v[i + 1, :]);
xlabel('x [m]'); ylabel('y [m]'); title("Car positions (Motion 1)"); axis('equal');
plot(r[:, 0], r[:, 1], '-k');
show();
xlabel('t [s]'); ylabel('v [m/s]'); title("Change of velocity in relation to the time of the car's movement");
plot(t, n_v, '-k');
show();
```

From the graph below, we can conclude that the car enters the cycle or ramp in the 4th second, where it loses speed at the top and begins to fall (4s - 5.5s). Then there are time intervals, each of which is shorter than the previous one. We can see an increase in velocity, that is, a drop, and then a short interval where the speed suddenly drops to zero and almost instantly jumps close to the old speed, which may indicate bouncing off the ground. The maximum acceleration in this case is at the moment immediately after touching the ground, which confirms the graph of the change of velovity, since, at that moment, the derivative is the highest.
